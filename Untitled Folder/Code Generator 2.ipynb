```python
import json
```

```python
with open("./complimentary_card.json") as file:
    complimentary_card_mapping = json.load(file)
```

```python
complimentary_card_mapping
```

```python
print("Map<CardType, Map<CardType, CardBenefits>> complimentaryCardMap = new HashMap<>();")
print("Map<CardType, CardBenefits> cardBenefitsMap;")
print("CardBenefits cardBenefits;")
print("Map<PurchaseCategory, Float> purchaseBenefitMap;")

for card in complimentary_card_mapping:  
    print("// Mapping the CardType to the benefits it offers")
    print('cardBenefitsMap = new HashMap<>();')
    print()
    
    for complimentary_card in complimentary_card_mapping[card]:
        print("// Preparing the CardBenefit class")
        print('cardBenefits = new CardBenefits();')
        print('cardBenefits.setCreditScore(' + str(complimentary_card[2]) + ');')
        print()
        print("// Mapping the Purchase Benefits to it's Category \n")
        print('purchaseBenefitMap = new HashMap<>();')
        for category in purchase_category_dict:
            value = complimentary_card[1][purchase_category_dict[category]]
            print('purchaseBenefitMap.put(PurchaseCategory.' + category + ', ' + str(value) + 'F);')
        print()
        print('cardBenefits.setPurchaseBenefitsMap(purchaseBenefitMap);')
        print('\n// Fitting in the cardBenefits with the card type \n')
        print('cardBenefitsMap.put(CardType.convert("' + complimentary_card[0] + '"), cardBenefits);')
        print()
        

    print('// Finally fitting in the cardBenefitMap with the original card')
    print('complimentaryCardMap.put(CardType.convert("' + card + '"), cardBenefitsMap);')
print('return complimentaryCardMap;')
```

```python
purchase_category_dict = {
    "EDUCATION": 0,
    "ENTERTAINMENT": 1,
    "FOOD": 2,
    "GAS_TRANS": 3,
    "GROCERY_NET": 4,
    "GROCERY_POS": 5,
    "HEALTH": 6,
    "HOME": 7,
    "HOTEL": 8,
    "KIDS_PETS": 9,
    "MISC_NET": 10,
    "MISC_POS": 11,
    "PERSONAL": 12,
    "SHOP_NET": 13,
    "SHOP_POS": 14,
    "TRAVEL": 15
}
```
